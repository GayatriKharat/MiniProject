```python
# 1. Descriptive Analysis: Rating Distribution and Top 10 Most-Rated Movies
ratings = open(r"C:\Users\hp\OneDrive\Desktop\ml-1m\ml-1m\ratings.dat")
rating_distribution = {'High': 0, 'Medium': 0, 'Low': 0}
movieId_count = {}

for line in ratings:
    line = line.strip()
    user_id, movie_id, rating, _ = map(int, line.split('::'))

    if rating == 5:
        rating_distribution['High'] += 1
    elif rating in [3, 4]:
        rating_distribution['Medium'] += 1
    else:
        rating_distribution['Low'] += 1

    if movie_id in movieId_count:
        movieId_count[movie_id] += 1
    else:
        movieId_count[movie_id] = 1

ratings.close()

# Calculate percentages
total_ratings = sum(rating_distribution.values())
for category, count in rating_distribution.items():
    print(f"{category}: {round((count / total_ratings) * 100, 2)}%")

# Top 10 Most-Rated Movies
movies = open(r"C:\Users\hp\OneDrive\Desktop\ml-1m\ml-1m\movies.dat")
movieId_name = {int(line.split('::')[0]): line.split('::')[1] for line in movies}
movies.close()

top_movies = sorted(movieId_count.items(), key=lambda x: x[1], reverse=True)[:10]
print("\nTop 10 Most-Rated Movies:")
for movie_id, count in top_movies:
    print(f"{movieId_name[movie_id]} - {count} ratings")
```

```python
#2. Genre Insights
#Code:

#python
#Copy
#Edit
# Genre Insights: Most Frequently Rated Genres & Average Ratings by Genre
movies = open (r"C:\Users\hp\OneDrive\Desktop\ml-1m\ml-1m\movies.dat")
genres_count = {}
genre_rating_sum = {}
genre_rating_count = {}

for line in movies:
    line = line.strip()
    movie_id, title, genres = line.split('::')
    genres_list = genres.split('|')

    for genre in genres_list:
        genres_count[genre] = genres_count.get(genre, 0) + 1

movies.close()

# Average Ratings by Genre
ratings = open(r"C:\Users\hp\OneDrive\Desktop\ml-1m\ml-1m\ratings.dat")
for line in ratings:
    line = line.strip()
    user_id, movie_id, rating, _ = map(int, line.split('::'))
    movie_genres = [g for g in movieId_name[movie_id].split('|')]

    for genre in movie_genres:
        genre_rating_sum[genre] = genre_rating_sum.get(genre, 0) + rating
        genre_rating_count[genre] = genre_rating_count.get(genre, 0) + 1

ratings.close()

average_ratings = {genre: round(genre_rating_sum[genre] / genre_rating_count[genre], 2) for genre in genre_rating_sum}
print("\nMost Frequently Rated Genres:")
for genre, count in genres_count.items():
    print(f"{genre}: {count} ratings")

print("\nAverage Ratings by Genre:")
for genre, avg in average_ratings.items():
    print(f"{genre}: {avg}")
```

```python
# Rating Distribution by Demographics
users = open(r"C:\Users\hp\OneDrive\Desktop\ml-1m\ml-1m\users.dat")
user_data = {}

for line in users:
    user_id, gender, age, occupation, _ = line.strip().split('::')
    user_data[int(user_id)] = (gender, age, occupation)

users.close()

ratings = open(r"C:\Users\hp\OneDrive\Desktop\ml-1m\ml-1m\ratings.dat")
demographic_ratings = {}

for line in ratings:
    user_id, movie_id, rating, _ = map(int, line.strip().split('::'))
    gender, age, occupation = user_data[user_id]

    key = (gender, age, occupation)
    if key not in demographic_ratings:
        demographic_ratings[key] = []
    demographic_ratings[key].append(rating)

ratings.close()

# Analyze and Print Ratings by Demographics
print("\nRatings by Demographics:")
for key, ratings_list in demographic_ratings.items():
    avg_rating = round(sum(ratings_list) / len(ratings_list), 2)
    print(f"Demographic {key}: Average Rating = {avg_rating}")
```

```python
 #5 Top Performers
movies_file = "movies.dat"
ratings_file = "ratings.dat"

movie_ratings = {}
movie_rating_counts = {}
with open(ratings_file, "r") as file:
    for line in file:
        movie_id = line.strip().split("::")[1]
        rating = line.strip().split("::")[2]
        movie_id = int(movie_id)
        rating = float(rating)

        if movie_id in movie_ratings:
            movie_ratings[movie_id].append(rating)
            movie_rating_counts[movie_id] += 1
        else:
            movie_ratings[movie_id] = [rating]
            movie_rating_counts[movie_id] = 1

min_ratings = 500
average_ratings = {
    movie_id: sum(ratings) / len(ratings)
    for movie_id, ratings in movie_ratings.items()
    if movie_rating_counts[movie_id] >= min_ratings
}
top_movies = sorted(average_ratings.items(), key=lambda x: x[1], reverse=True)

movie_details = {}
with open(movies_file, "r", encoding="ISO-8859-1") as file:
    for line in file:
        movie_id, title, genres = line.strip().split("::")
        movie_details[int(movie_id)] = {"title": title, "genres": genres}

print("Top Performers (Highest Average Ratings):")
for movie_id, avg_rating in top_movies[:10]:  # Top 10
    details = movie_details[movie_id]
    print(f"{details['title']} ({details['genres']}): {avg_rating:.2f}")
```

```python
#6 Exploring Long Tail
long_tail_movies = [movie_id for movie_id, count in movie_rating_counts.items() if count < 20]
popular_movies = [movie_id for movie_id, count in movie_rating_counts.items() if count >= 100]
long_tail_genres = {}
popular_genres = {}

for movie_id in long_tail_movies:
    genres = movie_details[movie_id]["genres"].split("|")
    for genre in genres:
        if genre in long_tail_genres:
            long_tail_genres[genre] += 1
        else:
            long_tail_genres[genre] = 1

for movie_id in popular_movies:
    genres = movie_details[movie_id]["genres"].split("|")
    for genre in genres:
        if genre in popular_genres:
            popular_genres[genre] += 1
        else:
            popular_genres[genre] = 1

print("\nCharacteristics of Long-Tail Movies (Genres):")
for genre, count in sorted(long_tail_genres.items(), key=lambda x: x[1], reverse=True):
    print(f"{genre}: {count}")

print("\nCharacteristics of Popular Movies (Genres):")
for genre, count in sorted(popular_genres.items(), key=lambda x: x[1], reverse=True):
    print(f"{genre}: {count}")
```
